```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_breast_cancer

#  
data = load_breast_cancer()
X = data.data
y = data.target
feature_names = data.feature_names

#  DataFrame     
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

#   
mean_radius = df['mean radius']
mean_texture = df['mean texture']
target = df['target']

#   
plt.figure(figsize=(10, 6))
scatter = plt.scatter(mean_radius, mean_texture, c=target, cmap='viridis', edgecolor='k', s=100)

#  
plt.legend(*scatter.legend_elements(), title="Classes", loc='upper right', labels=['Malignant', 'Benign'])

#      
plt.title(' :   ')
plt.xlabel(' ')
plt.ylabel(' ')

#  
plt.grid()
plt.show()
```
