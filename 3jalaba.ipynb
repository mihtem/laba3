```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_breast_cancer

# Загружаем данные
data = load_breast_cancer()
X = data.data
y = data.target
feature_names = data.feature_names

# Создаем DataFrame для удобства работы с данными
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

# Выбираем нужные признаки
mean_radius = df['mean radius']
mean_texture = df['mean texture']
target = df['target']

# Создаем диаграмму рассеяния
plt.figure(figsize=(10, 6))
scatter = plt.scatter(mean_radius, mean_texture, c=target, cmap='viridis', edgecolor='k', s=100)

# Добавляем легенду
plt.legend(*scatter.legend_elements(), title="Classes", loc='upper right', labels=['Malignant', 'Benign'])

# Добавляем подписи к осям и заголовок
plt.title('Диаграмма рассеяния: Рак молочной железы')
plt.xlabel('Среднний радиус')
plt.ylabel('Средняя текстура')

# Показываем график
plt.grid()
plt.show()
```
